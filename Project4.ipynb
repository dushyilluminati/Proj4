````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
grouped_df = df.groupby('Project_Phase')['Actual_Variance'].agg([
    ('Avg_Variance', 'mean'),
    ('Max_Positive_Variance', lambda x: x[x > 0].max()),  # Maximum positive variance
    ('Max_Negative_Variance', lambda x: x[x < 0].min()),  # Maximum negative variance
    ('Median_Variance', 'median')
]).reset_index()
````

Marking extreme values and measures of central tendencies.

````python
from pulp import LpMinimize, LpProblem, LpVariable, lpSum

# Define the problem
model = LpProblem(name="cost-optimization", sense=LpMinimize)

# Define variables for each project to adjust allocation
variables = {row["Project_ID"]: LpVariable(name=row["Project_ID"], lowBound=0) for _, row in df.iterrows()}

# Objective: Minimize total adjustments
model += lpSum(variables[proj_id] for proj_id in variables), "Minimize_Adjustments"

# Constraints
for _, row in df.iterrows():
    # Ensure adjusted budget covers actual expenditure
    model += variables[row["Project_ID"]] >= row["Actual_Expenditure"], f"{row['Project_ID']}_Budget_Constraint"

# Solve the problem
model.solve()

# Display Results
results = []
for var in model.variables():
    results.append({"Project_ID": var.name, "Optimized_Budget": var.value()})

# Convert to DataFrame
optimized_df = pd.DataFrame(results)
print(optimized_df)
````
