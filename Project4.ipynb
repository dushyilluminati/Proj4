````python
df=pd.read_csv('Dataset - Project 4.csv')
````

````python
df['Actual_Variance'] = df['Allocated_Budget'] - df['Actual_Expenditure']
````

The enlisted Variance column has wrong values, hence created new column


````python
df
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
grouped_df = df.groupby('Project_Phase')['Actual_Variance'].agg([
    ('Avg_Variance', 'mean'),
    ('Max_Positive_Variance', lambda x: x[x > 0].max()),  # Maximum positive variance
    ('Max_Negative_Variance', lambda x: x[x < 0].min()),  # Maximum negative variance
    ('Median_Variance', 'median')
]).reset_index()
````

Marking extreme values and measures of central tendencies.


````python
print(grouped_df)
````

````python
plt.figure(figsize=(10, 5))
sns.barplot(x='Project_Phase', y='Actual_Variance', data=df)
plt.ylabel('Actual Variance')
plt.xlabel('Project Phase')
plt.show()
````

Planning- Projects are over budget due to consistent negative variance
Execution- Projects are fairing well on budget but will require new strategy to allocate better budget to avoid over spending
Closure- More projects are over budget with a larger count than planning.
